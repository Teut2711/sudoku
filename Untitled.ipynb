```python
points = []
def point_put(event, x, y, flags, params):
    global points
    if event == cv2.EVENT_LBUTTONDOWN:
        points += [(x,y)]
cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.setMouseCallback('image',point_put)
while(1):
    cv2.imshow('image',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
X = list()
for i in [cv2.imread(os.path.join("sudoku images", i), 0) for i in df.Image]:
    
    X.append(cv2.resize(i,(640, 480)))
```

```python
y = [np.array(eval(i)) for i in df.Points]
```

```python
X = np.array(X).reshape(-1, 480, 640, 1)
```

```python
y = np.array(y )
```

```python
X.shape, y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Conv2D(10,(3, 4), activation='linear',  input_shape=(480, 640, 1)))

model.add(Conv2D(20, (3, 4), activation='linear'))

model.add(Conv2D(20, (3, 4), activation='linear'))
model.add(Conv2D(20, (3, 4), activation='linear'))
model.add(MaxPool2D((2, 3)))
model.add(Conv2D(20, (3, 4), activation='linear'))
model.add(MaxPool2D((3, 4)))
model.add(Conv2D(10, (3, 4), activation='linear'))
model.add(MaxPool2D((2, 4)))
model.add(Conv2D(40, (4, 5), activation='linear'))
model.add(MaxPool2D((2, 4)))

model.add(Conv2D(1, (2, 1 ), activation='linear'))
model.compile(loss='mean_squared_error', optimizer=Adam(.5) )
```

```python
model.summary()
```

```python
history = model.fit(x=X, y=y, epochs=50,batch_size=10)
```
