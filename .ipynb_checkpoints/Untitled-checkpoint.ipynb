```python
import tensorflow as tf
from PIL import Image
import cv2 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data.csv", index_col=0) 
```

```python
import os
images = sorted(list(filter(lambda x:(x.endswith("png") or x.endswith("jpg")) and x not in df.Image.tolist(), os.listdir("sudoku images"))), key=len)
```

```python
i = len(df) 
```

```python
img = cv2.imread(os.path.join("sudoku images", images[i]), 0)
if img.shape != (480, 640):
    img = cv2.resize(img, (480, 640))
```

```python
df
```

```python
%matplotlib notebook
plt.imshow(img)
```

```python
points = []
def point_put(event, x, y, flags, params):
    global points
    if event == cv2.EVENT_LBUTTONDOWN:
        points += [(x,y)]
cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.setMouseCallback('image',point_put)
while(1):
    cv2.imshow('image',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
df = df.append({"Image":images[i], "Points":str(points)}, ignore_index=True)
```
